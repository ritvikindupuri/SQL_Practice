## Question 2

```python
%%sql
-- Joining Salaries and People on their common playerID
SELECT * FROM Salaries
JOIN People ON Salaries.playerID = People.playerID
LIMIT 5;
```

```python
%%sql
-- Creating a temporary table 'joined_data' with the combined info
WITH joined_data AS (
    SELECT * FROM Salaries JOIN People ON People.playerID = Salaries.playerID
)
-- Querying the temporary table to get the average salary by weight
SELECT
    weight,
    AVG(salary) AS avg_salary
FROM joined_data
GROUP BY weight
ORDER BY avg_salary DESC
LIMIT 10;
```

The JOIN command is fundamental for leveraging the power of a relational database. My first query successfully merged player biographical data (like names from People) with their financial data (salary from Salaries). The second query demonstrates a more practical application: by joining the tables, I was able to analyze the relationship between a physical attribute (weight) and compensation. The results, showing the average salary for different weight groups, could be the first step in analyzing if there's a correlation between a player's physique and their earnings. Using a WITH clause made the final query much more readable and manageable.



## Question 3

```python
%%sql
-- Joining People and Batting to link names to stats
WITH joined_data AS (
    SELECT * FROM People JOIN Batting ON People.playerID = Batting.playerID
)
-- Summing all runs ('R') for the player named Babe Ruth
SELECT SUM(R)
FROM joined_data
WHERE nameFirst = 'Babe' AND nameLast = 'Ruth';
```

```python
%%sql
-- Joining People and Batting
WITH joined_data AS (
    SELECT * FROM People JOIN Batting ON People.playerID = Batting.playerID
)
-- Filtering for players born in the 1980s
-- Then calculating the average runs, grouped and ordered by their birth year
SELECT
    birthYear,
    AVG(R) as avg_runs
FROM joined_data
WHERE birthYear >= 1980 AND birthYear <= 1989
GROUP BY birthYear
ORDER BY birthYear;
```

These queries show how joining tables allows for deep, specific analysis. The first query successfully calculated Babe Ruth's career runs (2174), demonstrating how to aggregate data for a single entity across many years of records. The second query provided a demographic snapshot, revealing trends in the average performance of players born throughout the 1980s. This type of analysis could be used to compare different generations of players or identify peak performance years for a certain cohort.

## Question 4

```python
%%sql
-- Joining People and Batting to get names and stats
-- Summing career runs (R) for each player
SELECT
    p.nameFirst,
    p.nameLast,
    SUM(b.R) AS total_runs
FROM People p
JOIN Batting b ON p.playerID = b.playerID
GROUP BY p.playerID
ORDER BY total_runs DESC
LIMIT 10;
```

```python
%%sql
-- Joining People and Batting to get names and stats
-- Summing career hits (H) for each player
SELECT
    p.nameFirst,
    p.nameLast,
    SUM(b.H) AS total_hits
FROM People p
JOIN Batting b ON p.playerID = b.playerID
GROUP BY p.playerID
ORDER BY total_hits DESC
LIMIT 10;
```

```python
%%sql
-- Using the provided WITH clause to join CollegePlaying and Batting for Purdue players
WITH joined_data AS (
    SELECT CollegePlaying.playerID, Batting.R
    FROM CollegePlaying
    JOIN Batting ON CollegePlaying.playerID = Batting.playerID
    WHERE CollegePlaying.schoolID = 'purdue'
)
-- Summing the career runs for each Purdue player
SELECT
    playerID,
    SUM(R) AS total_runs
FROM joined_data
GROUP BY playerID
ORDER BY total_runs DESC
LIMIT 10;
```

```python
%%sql
-- Joining CollegePlaying and Batting
-- Calculating the average runs and grouping by school
SELECT
    cp.schoolID,
    AVG(b.R) as avg_runs
FROM CollegePlaying cp
JOIN Batting b ON cp.playerID = b.playerID
GROUP BY cp.schoolID
ORDER BY avg_runs DESC
LIMIT 10;
```
